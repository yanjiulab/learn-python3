```python
from collections import Iterable, Iterator
isinstance((), Iterable)
```

```python
def fib(max):
    n, a, b = 0, 0, 1
    while n < max:
        yield b
        a, b = b, a + b
        n = n + 1
    return 'done'
```

```python
for i in fib(10):
    print(i)
```

```python
isinstance(fib(10), Iterator)
```
